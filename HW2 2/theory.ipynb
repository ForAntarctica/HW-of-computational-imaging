# 作业1

题目中给出了模型：

- **观测数据**：
  $$
  Y = (Y_1, Y_2, \dots, Y_n) \sim \text{Exp}(\lambda)
  $$

- **先验分布**：
  $\lambda$ 的概率密度函数为：
  $$
  p(\lambda) = 
  \begin{cases}
    \alpha e^{-\alpha \lambda}, & \lambda \geq 0 \\
    0, & \lambda < 0
  \end{cases}
  $$

## 1. MMSE Estimator
讲义中有，估计器为：
$$
E(\lambda |Y = y)
$$
现在求后验分布$p(\lambda | Y = y)$，根据贝叶斯定理：
$$
p(\lambda | Y = y) = \frac{p(Y = y | \lambda) p(\lambda)}{p(Y = y)}
$$
其中，$p(Y = y)$ 是归一化常数，与 $\lambda$ 无关，因此在推导中可以忽略。

### 1.1 似然函数 $p(Y = y | \lambda)$

由于 $Y_1, Y_2, \dots, Y_n$ 独立同分布，似然函数为：
$$
p(Y = y | \lambda) = \prod_{i=1}^{n} p(Y_i = y_i | \lambda) = \prod_{i=1}^{n} \lambda e^{-\lambda y_i} = \lambda^n e^{-\lambda \sum_{i=1}^{n} y_i}
$$
当任意 $y_i < 0$ 时，$p(Y = y | \lambda) = 0$。

### 1.2 先验分布 $p(\lambda)$

$$
p(\lambda) = 
\begin{cases}
  \alpha e^{-\alpha \lambda}, & \lambda \geq 0 \\
  0, & \lambda < 0
\end{cases}
$$

### 1.3 后验分布的非归一化形式

$$
p(\lambda | Y = y) \propto p(Y = y | \lambda) p(\lambda) = \lambda^n e^{-\lambda \sum_{i=1}^{n} y_i} \cdot \alpha e^{-\alpha \lambda} = \alpha \lambda^n e^{-\lambda \left( \sum_{i=1}^{n} y_i + \alpha \right)}
$$
且 $\lambda \geq 0$，否则 $p(\lambda | Y = y) = 0$。

### 1.4 求后验概率的均值

后验分布 $p(\lambda | Y = y)$ 的形式类似于 Gamma 分布。标准的 Gamma 分布 $\text{Gamma}(k, \theta)$ 的概率密度函数为：
$$
p(x) = \frac{\theta^k}{\Gamma(k)} x^{k-1} e^{-\theta x}, \quad x \geq 0
$$
根据前文的结果，可以得到：
$$
k = n + 1 \quad \text{（形状参数）}, \quad \theta = \sum_{i=1}^{n} y_i + \alpha \quad \text{（率参数）}
$$
因此，后验分布 $p(\lambda | Y = y)$ 是一个形状参数为 $k = n + 1$、率参数为 $\theta = \sum_{i=1}^{n} y_i + \alpha$ 的 Gamma 分布：
$$
\lambda | Y = y \sim \text{Gamma}(n + 1, \sum_{i=1}^{n} y_i + \alpha)
$$

## 2. MAP Estimator

$$
\hat{\lambda}_{MAP} = \arg \max \limits_x \{\log p(Y= y|\lambda)+\log(\lambda)\}
$$
通过上面的结果有：
$$
\hat{\lambda}_{MAP} = \arg \max \limits_x \{n\log \lambda -(\sum_{i=1}^{n} y_i + \alpha)\lambda +\log(\alpha)\}
$$

**利用导数求极值**：
令对$\lambda$的导数等于0:
$$
\frac{n}{\lambda} = \sum_{i=1}^{n} y_i + \alpha
$$
$$
\lambda = \frac{n}{\sum_{i=1}^{n} y_i + \alpha}
$$
**MAP Estimator 的形式**：
$$
\hat{\lambda}_{MAP}(Y) = \frac{n}{\sum_{i=1}^{n} Y_i + \alpha}
$$

### 3 MMSE Estimator

**定义**：MMSE 估计器是后验分布的期望值，即最小化均方误差的估计量。

对于 Gamma 分布 $\text{Gamma}(k, \theta)$，其期望值为：
$$
\mathbb{E}[\lambda | Y = y] = \frac{k}{\theta}
$$
因此，对于后验分布 $\lambda | Y = y \sim \text{Gamma}(n + 1, \sum y_i + \alpha)$，
$$
\hat{\lambda}_{MMSE} = \frac{n + 1}{\sum_{i=1}^{n} y_i + \alpha}
$$

**Estimator 的形式**：
$$
\hat{\lambda}_{MMSE}(Y) = \frac{n + 1}{\sum_{i=1}^{n} Y_i + \alpha}
$$

### 2.3 最小均绝对误差估计器 (MMAE Estimator)

**定义**：MMAE 估计器是后验分布的中位数，即最小化均绝对误差的估计量。

对于 Gamma 分布 $\text{Gamma}(k, \theta)$，其中位数通常没有简单的解析表达式，需通过数值方法或查表获得。

**Estimator 的形式**：
$$
\hat{\lambda}_{MMAE}(Y) = \text{median}(\lambda | Y = y)
$$
具体计算时，可以利用数值算法如二分法、牛顿法等，或使用统计软件中的相关函数来估计 Gamma 分布的中位数。

## 3. 总结

在给定观测数据 $Y = (Y_1, \dots, Y_n)$ 服从参数为 $\lambda$ 的指数分布，且 $\lambda$ 服从参数为 $\alpha$ 的指数先验分布的情况下：

- **MAP 估计器**：
  $$
  \hat{\lambda}_{MAP}(Y) = \frac{n}{\sum_{i=1}^{n} Y_i + \alpha}
  $$

- **MMSE 估计器**：
  $$
  \hat{\lambda}_{MMSE}(Y) = \frac{n + 1}{\sum_{i=1}^{n} Y_i + \alpha}
  $$

- **MMAE 估计器**：
  $$
  \hat{\lambda}_{MMAE}(Y) = \text{median}(\lambda | Y = y)
  $$
  其中，中位数需通过数值方法计算。

这些估计器均基于后验分布 $\lambda | Y = y \sim \text{Gamma}(n + 1, \sum y_i + \alpha)$，并严格遵守定义域条件，即 $Y_i \geq 0$ 且 $\lambda \geq 0$。
